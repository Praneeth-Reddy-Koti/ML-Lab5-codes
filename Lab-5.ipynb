```python
import math
no_of_ephocs=0

def step(x):
    y = math.modf(x)
    return y[1]


def epoch(bias, w1, w2, x1, x2, t, alpha, no_of_ephocs):
    flag = False

    for i in range(4):
        yin = float(w1 * x1[i] + x2[i] * w2 + bias)

        if step(yin) == t[i]:
            continue
        else:
            print("Weights are changing")
            w1 = w1 + alpha * (t[i] - step(yin)) * x1[i]
            w2 = w2 + alpha * (t[i] - step(yin)) * x2[i]
            bias = bias + alpha * (t[i] - step(yin))
            flag = True

        print(f"Weights in {i + 1} iteration")
        print(f"W1 = {w1}")
        print(f"W2 = {w2}")
        print(f"Bias = {bias}")

    if flag:
        print("Weights are changed in the previous epoch")
        print(f"epoch {no_of_ephocs}")
        epoch(bias, w1, w2, x1, x2, t, alpha, no_of_ephocs + 1)
    else:
        print("Weights are not changed ")
        print("Final weights are: ")
        print(f"W1 = {w1}")
        print(f"W2 = {w2}")
        print(f"Bias = {bias}")


bias = 10
w1 = 0.2
w2 = -0.75
alpha = 0.05
x1 = [0, 0, 1, 1]
x2 = [0, 1, 0, 1]
t = [0, 0, 0, 1]

epoch(bias, w1, w2, x1, x2, t, alpha, 1)
```
